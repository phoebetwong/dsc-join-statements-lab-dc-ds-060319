## Connecting to the Database
Import the necessary packages and connect to the database **data.sqlite**.

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('data.sqlite')
c = conn.cursor()
```

## Display the names of all the employees in Boston.
Hint: join the employees and offices tables.

```python
c.execute('''select lastName, firstName from employees join offices using(officeCode) where city = 'Boston';''').fetchall()
```

## Do any offices have no employees?
Hint: Combine the employees and offices tables and use a group by.

```python
c.execute('''select city, count(employeeNumber) as num_employees from offices left join employees using(officeCode) 
group by (officeCode);''').fetchall()


```
